```python
import os
import shutil
from photos import inspect_library, process_library

ORIGINAL_LIBRARY_DIR = "original-photos"
PROCESSED_LIBRARY_DIR = "processed-photos"
```

```python
prior_metadata = inspect_library(ORIGINAL_LIBRARY_DIR)
prior_metadata
```

```python
if os.path.exists(PROCESSED_LIBRARY_DIR):
    shutil.rmtree(PROCESSED_LIBRARY_DIR)

process_library(ORIGINAL_LIBRARY_DIR, PROCESSED_LIBRARY_DIR)
```

```python
posterior_metadata = inspect_library(PROCESSED_LIBRARY_DIR)
total_size_reduction = (
    prior_metadata["image_size_mb"].sum() - posterior_metadata["image_size_mb"].sum()
)
percentage_reduction = (
    total_size_reduction / prior_metadata["image_size_mb"].sum() * 100
)
reduction_per_image = total_size_reduction / len(prior_metadata)
print(
    f"Reduced {total_size_reduction:.2f} MB ({percentage_reduction:.2f}%, {reduction_per_image:.2f} MB per image) in total image size."
)
print(
    f"Expected reduction of total library of size 13GB: {13 * percentage_reduction / 100:.2f} GB"
)
```
